```r
library(mlr3)
library(mlr3learners)
library(mlr3pipelines)
library(mlr3tuning)
library(data.table)
library(future)
library(future.apply)
library(neuralnet)
library(kknn)
library(ggplot2)
library(e1071)
library(mltools)
```

```r
data <- read.csv("data60.csv", header = TRUE)
```

```r
data[data == '?'] <- NA
data <- as.data.frame(lapply(data, as.numeric))
data <- na.omit(data)
binary_cols = colnames(data[,grepl("binary",colnames(data))])
binary_cols = colnames(data[,binary_cols][,which(apply(data[,binary_cols], 2, function(x) length(unique(x))) <= 2)])
data[,binary_cols] <- as.data.frame(lapply(data[,binary_cols], as.factor))    

n_targets <- 33
target_cols <- colnames(data[,(ncol(data) - n_targets + 1):ncol(data)])
data$target <- rowSums(data[,target_cols])
data <- data[,!(colnames(data) %in% c(target_cols, "ut_ms"))]
set.seed(42)
data <- data[sample(1:nrow(data), size = 10000, replace = FALSE),]
```

```r
print(paste0("Number rows: ", nrow(data)))
print(paste0("Number features: ", ncol(data)))
```

# KNN - Regression

```r
task <- TaskRegr$new(id="data_task", data, target = "target")
split <- partition(task, ratio = 0.8)
train_task <- TaskRegr$new(id="train_task", data[split$train,], target = "target")
test_task <- TaskRegr$new(id="test_task", data[split$test,], target = "target")

gr_knn <- po("scale", affect_columns = selector_type("numeric"), id = "scaling") %>>% 
        po("encode", method = "one-hot", affect_columns = selector_type("factor"), id = "binary_enc") %>>%
        lrn("regr.kknn")

grl_knn = GraphLearner$new(gr_knn)

tuner_params <- ParamSet$new(
  list(
    ParamInt$new("regr.kknn.k", lower = 1, upper = 20),
    ParamInt$new("regr.kknn.distance", lower = 1, upper = 2),
    ParamFct$new("regr.kknn.kernel", levels = c("rectangular", "optimal"))
  )
)

design <- expand.grid(regr.kknn.k = seq(1, 20, 2),
                      regr.kknn.distance = c(1, 2),
                      regr.kknn.kernel = c("rectangular", "optimal"))

design = data.table(regr.kknn.k = design$regr.kknn.k,
                    regr.kknn.distance = design$regr.kknn.distance,
                    regr.kknn.kernel = as.character(design$regr.kknn.kernel))

tuner = tnr("design_points", design = design)

instance_knn = ti(
  task = train_task,
  learner = grl_knn,
  resampling = rsmp("cv", folds = 5),
  measures = msr("regr.rmse"),
  terminator = trm("none"),
  search_space = tuner_params
)
```

```r
plan(multisession, workers = 7)
set.seed(42, "L'Ecuyer-CMRG")
tuner$optimize(instance_knn)
plan(sequential)
```

```r
d <- instance_knn$archive$data[,c("regr.kknn.k", "regr.kknn.distance", "regr.kknn.kernel", "regr.rmse")]
d$regr.kknn.distance <- factor(d$regr.kknn.distance)
d$regr.kknn.kernel <- factor(d$regr.kknn.kernel)
```

```r
options(repr.plot.width=15, repr.plot.height=8)
ggplot(data = d, mapping = aes(x = regr.kknn.k, y = regr.rmse, color = regr.kknn.distance, shape = regr.kknn.kernel)) + 
    geom_point(size = 6) +
    geom_line() +
    theme_bw() +
    theme(legend.text = element_text(size = 12),
          legend.key.size = unit(2, "cm"), 
          axis.title = element_text(size = 16),
          axis.text = element_text(size = 16))
```

```r
model <- po("scale", affect_columns = selector_type("numeric")) %>>% 
    po("encode", method = "one-hot", affect_columns = selector_type("factor")) %>>%
    lrn("regr.kknn", k = 17, distance = 1, kernel = "optimal")
model$train(train_task)
```

```r
pred_test_knn <- model$predict(test_task)
```

```r
rmse <- function(truth, pred){
    round(sqrt(mean((pred - truth)^2)), 4)
}
cat("KNN test RMSE =", rmse(pred_test_knn$regr.kknn.output$truth, pred_test_knn$regr.kknn.output$response))
```

```r
pred_train <- model$predict(train_task)
cat("KNN train RMSE =", rmse(pred_train$regr.kknn.output$truth, pred_train$regr.kknn.output$response))
```

# Neural Network - Regression

```r
data1 <- as.data.frame(lapply(data, as.numeric))
neuralnet(formula = target ~ ., 
          data = data1[split$train,],
          hidden = c(10, 10), 
          rep = 1,
          linear.output = TRUE) -> nn_model
```

```r
pred_test_nn <- predict(nn_model, newdata = data1[split$test,])
```

```r
rmse <- function(truth, pred){
    round(sqrt(mean((pred - truth)^2)), 4)
}
cat("NN test RMSE =", rmse(truth = data1$target[split$test], pred = pred_test_nn))
```

```r
pred_train <- predict(nn_model, newdata = data1[split$train,])
cat("NN train RMSE =", rmse(truth = data1$target[split$train], pred_train))
```

# SVM

```r
svm_model <- svm(target ~ .,
                data = data[split$train,],
                kernel = "radial",
                cost = 1, 
                scale = FALSE)
```

```r
pred_test_svm <- predict(svm_model, newdata = data[split$test,])
rmse <- function(truth, pred){
    round(sqrt(mean((pred - truth)^2)), 4)
}
cat("SVM test RMSE =", rmse(truth = data$target[split$test], pred = pred_test_svm))
```

```r
pred <- predict(svm_model, newdata = data[split$train,])
cat("SVM train RMSE =", rmse(truth = data$target[split$train], pred = pred))
```
