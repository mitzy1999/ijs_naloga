```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates
import seaborn as sns
import os

from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor

from sklearn.metrics import mean_squared_error
from sklearn.metrics import make_scorer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_selector as selector
```

```python
DATA_DIR = "~/Documents/IJS/naloga1/"
data = pd.read_csv(os.path.join(DATA_DIR, "data60.csv"), low_memory=False)
```

```python
print(f"Data shape (resolution 60min) : {data.shape}\nNumber of rows {data.shape[0]}\nNumber of columns {data.shape[1]}")
```

```python
# replace ? for NAN
data = data.replace('?', np.nan)
# Delete rows with missing values
data = data.dropna()
print(f"Data shape (resolution 60min) : {data.shape}\nNumber of rows {data.shape[0]}\nNumber of columns {data.shape[1]}")
```

```python
# Select data sample (10k rows) and seperate target and features
data = data.astype(float)
binary_cols = np.array([col for col in data.columns if "binary" in col])
binary_cols = binary_cols[np.where((data[binary_cols].nunique() <= 2) == True)[0]]
data[binary_cols] = data[binary_cols].astype("O")
data = data.sample(n = 10000, random_state = 42)
target_columns = data.iloc[:,-33:].columns
target = data.loc[:, target_columns].astype(float).sum(axis=1)
target_max = data.loc[:, target_columns].astype(float).max(axis=1)
data.drop(columns=target_columns, inplace=True)
print(f"Data shape (resolution 60min) : {data.shape}\nNumber of rows {data.shape[0]}\nNumber of columns {data.shape[1]}")
```

```python
#save df for orange
orange_data = pd.concat([data, pd.DataFrame(target, columns=["target"])], axis=1)
orange_data.to_csv(os.path.join(DATA_DIR, "orange_data.csv"), index = False)
del orange_data
```

```python
df_plot = pd.DataFrame({
    "time" : data["ut_ms"],
    "target" : target
})
df_plot = df_plot.sort_index()
df_plot["time"] = [datetime.fromtimestamp(x/1000.0).strftime('%Y-%m-%d') for x in df_plot["time"]]
df_plot["time"] = pd.to_datetime(df_plot["time"])
df_plot = df_plot.set_index("time")

fig, ax = plt.subplots(figsize=(15, 5))

ax.plot(df_plot.index, df_plot["target"])

ax.xaxis.set_major_locator(mdates.YearLocator(base = 1, month = 1, day = 1)) # Set the major locator to months
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Format the date as you like
```

```python
df_plot = pd.DataFrame({
    "time" : data["ut_ms"],
    "target" : target_max
})
df_plot = df_plot.sort_index()
df_plot["time"] = [datetime.fromtimestamp(x/1000.0).strftime('%Y-%m-%d') for x in df_plot["time"]]
df_plot["time"] = pd.to_datetime(df_plot["time"])
df_plot = df_plot.set_index("time")

fig, ax = plt.subplots(figsize=(15, 5))

ax.plot(df_plot.index, df_plot["target"])

ax.xaxis.set_major_locator(mdates.YearLocator(base = 1, month = 1, day = 1)) # Set the major locator to months
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Format the date as you like
```

# KNN - Regression

```python
data.drop(columns=["ut_ms"], inplace=True)
del df_plot
X_train, X_test, y_train, y_test = train_test_split(data, target, train_size=0.8, random_state=42)
```

```python
def rmse(predict, actual):
    predict, actual = np.array(predict), np.array(actual)
    return np.sqrt(np.mean((predict - actual)**2))

rmse_score = make_scorer(rmse, greater_is_better = False)

model = KNeighborsRegressor()

numeric_transformer = StandardScaler()

categorical_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, selector(dtype_exclude="category")),
        ("cat", categorical_transformer, selector(dtype_include="category"))
    ]
)

pipe = Pipeline(steps=[('preprocessor', preprocessor), ("regression", model)])

parameters = {"regression__n_neighbors" : np.arange(1, 20, 2), 
              "regression__weights" : ["uniform", "distance"],
              "regression__p" : [1, 2]}

knn_regresor = GridSearchCV(pipe, parameters, scoring=rmse_score)
knn_regresor.fit(X_train, y_train)
print("Best parameter (CV score=%0.3f):" % knn_regresor.best_score_)
print(knn_regresor.best_params_)
```

```python
plt.figure(figsize=(10, 6))

d = pd.DataFrame({
    'param_n_neighbors' : knn_regresor.cv_results_['param_regression__n_neighbors'].data,
    'param_p' : knn_regresor.cv_results_['param_regression__p'].data,
    'param_weights' : knn_regresor.cv_results_['param_regression__weights'].data,
    'mean_test_score' : -1 * knn_regresor.cv_results_['mean_test_score']
})

sns.lineplot(data=d, 
             x="param_n_neighbors", 
             y="mean_test_score", 
             hue="param_p", 
             style="param_weights", 
             markers=["o", "o"], 
             palette=['red', 'blue'])
```

```python
print(f"Test RMSE : {round(-1 * knn_regresor.score(X_test, y_test), 4)}")
```

```python
print(f"Train RMSE : {round(-1 * knn_regresor.score(X_train, y_train), 4)}")
```

# Neural networks

```python
model = MLPRegressor(random_state=42, max_iter=400)

numeric_transformer = StandardScaler()

categorical_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, selector(dtype_exclude="category")),
        ("cat", categorical_transformer, selector(dtype_include="category"))
    ]
)

pipe = Pipeline(steps=[('preprocessor', preprocessor), ("regression", model)])

parameters = {"regression__hidden_layer_sizes" : [(10, 10), (100, 100)], 
              "regression__alpha" : [0.0001, 0.001],
              "regression__learning_rate_init" : [0.001, 0.01]}

nn_regresor = GridSearchCV(pipe, parameters, scoring=rmse_score)
nn_regresor.fit(X_train, y_train)
print("Best parameter (CV score=%0.3f):" % nn_regresor.best_score_)
print(nn_regresor.best_params_)
```

```python
plt.figure(figsize=(10, 6))

d = pd.DataFrame({
    'param_hidden_layer_sizes' : nn_regresor.cv_results_['param_regression__hidden_layer_sizes'].data,
    'param_alpha' : nn_regresor.cv_results_['param_regression__alpha'].data,
    'param_learning_rate' : nn_regresor.cv_results_['param_regression__learning_rate_init'].data,
    'mean_test_score' : -1 * nn_regresor.cv_results_['mean_test_score']
})

ax = sns.lineplot(data=d, 
             x="param_alpha", 
             y="mean_test_score", 
             hue="param_hidden_layer_sizes", 
             style="param_learning_rate", 
             markers=["o", "o"], 
             palette=['red', 'blue'])
sns.move_legend(ax, "lower center", bbox_to_anchor=(.5, 1), ncol=6, title=None, frameon=False)
```

```python
print(f"Test RMSE : {round(-1 * nn_regresor.score(X_test, y_test), 4)}")
```

```python
print(f"Train RMSE : {round(-1 * nn_regresor.score(X_train, y_train), 4)}")
```

# Regression trees

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.svm import SVR
from sklearn.model_selection import RandomizedSearchCV
```

```python
def rmse(predict, actual):
    predict, actual = np.array(predict), np.array(actual)
    return np.sqrt(np.mean((predict - actual)**2))

rmse_score = make_scorer(rmse, greater_is_better = False)

rt = DecisionTreeRegressor(random_state=42).fit(X_train, y_train)
print(f"Test RMSE : {round(rmse(rt.predict(X_test), y_test), 4)}")
```

```python
print(f"Train RMSE : {round(rmse(rt.predict(X_train), y_train), 4)}")
```

## Pre-pruning

```python
space = {"criterion" : ["squared_error", "friedman_mse", "absolute_error", "poisson"], 
         "max_depth" : list(np.arange(2, 10, step=1)),
         "min_samples_split" : list(np.arange(10, 100, step=10)),
         "min_samples_leaf" : list(np.arange(10, 100, step=10))}

rt = DecisionTreeRegressor(random_state=42)
rt_regresor = RandomizedSearchCV(rt, space,  n_iter=50, cv=5, scoring=rmse_score, random_state=42, n_jobs=-1)
rt_regresor.fit(X_train, y_train)
print("Best parameter (CV score=%0.3f):" % rt_regresor.best_score_)
print(rt_regresor.best_params_)
```

```python
print(f"Test RMSE : {round(-1 * rt_regresor.score(X_test, y_test), 4)}")
```

```python
print(f"Train RMSE : {round(-1 * rt_regresor.score(X_train, y_train), 4)}")
```

```python
plt.figure(figsize=(20, 8))
plot_tree(rt_regresor.best_estimator_)
plt.show()
```

# SVM - Regression

```python
model = SVR()

numeric_transformer = StandardScaler()

categorical_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, selector(dtype_exclude="category")),
        ("cat", categorical_transformer, selector(dtype_include="category"))
    ]
)

pipe = Pipeline(steps=[('preprocessor', preprocessor), ("regression", model)])

parameters = {"regression__kernel" : ["rbf", "poly"], 
              "regression__C" : [0.0001, 0.001, 0.01, 0.1, 1, 10, 100]}

svm_regresor = GridSearchCV(pipe, parameters, scoring=rmse_score, n_jobs=-1)
svm_regresor.fit(X_train, y_train)
print("Best parameter (CV score=%0.3f):" % svm_regresor.best_score_)
print(svm_regresor.best_params_)
```

```python
print(f"Test RMSE : {round(-1 * svm_regresor.score(X_test, y_test), 4)}")
```

```python
print(f"Train RMSE : {round(-1 * svm_regresor.score(X_train, y_train), 4)}")
```
