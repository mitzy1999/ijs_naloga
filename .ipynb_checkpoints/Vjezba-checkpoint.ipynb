```python
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 50)
y = np.sin(x) + np.random.randn(50) * 0.1
plt.plot(x, y, 'bo')
```

```python
model1 = DecisionTreeRegressor(max_depth=10).fit(x.reshape(-1, 1), y)
model2 = DecisionTreeRegressor(max_depth=2).fit(x.reshape(-1, 1), y)
```

```python
pred1 = model1.predict(x.reshape(-1, 1))
pred2 = model2.predict(x.reshape(-1, 1))
```

```python
plt.plot(x,pred1, 'g')
plt.plot(x,pred2, 'r')
plt.plot(x, y, 'bo')
```

```python
plt.figure(figsize=(10, 10))
tree.plot_tree(model2)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
tree.plot_tree(model1)
plt.show()
```
